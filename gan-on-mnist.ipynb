# GAN on MNIST (PyTorch)

```python
# ----- imports -----
import os, random, time
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms, utils as vutils
import matplotlib.pyplot as plt

# ----- setup -----
seed = 42                          # keep runs reproducible
random.seed(seed); torch.manual_seed(seed)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# hyperparams (kept simple)
IMG_SIZE, IMG_CH = 28, 1
Z_DIM = 100                        # noise size
G_HIDDEN, D_HIDDEN = 256, 256
BATCH_SIZE = 128
LR, BETAS = 2e-4, (0.5, 0.999)
EPOCHS = 200
CHECKPOINT_EPOCHS = {1, 5, 50, 100, 150, 200}  # snapshots to inspect
SAMPLES_DIR = "samples"; os.makedirs(SAMPLES_DIR, exist_ok=True)

# ----- data -----
# MNIST to [-1, 1] so Tanh in G matches it
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,), (0.5,))
])
train_set = datasets.MNIST(root="./data", train=True, transform=transform, download=True)
loader = DataLoader(train_set, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)

# ----- models -----
class Generator(nn.Module):
    # simple MLP generator (fast + clear)
    def __init__(self, zdim=Z_DIM):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(zdim, G_HIDDEN), nn.BatchNorm1d(G_HIDDEN), nn.ReLU(True),
            nn.Linear(G_HIDDEN, G_HIDDEN*2), nn.BatchNorm1d(G_HIDDEN*2), nn.ReLU(True),
            nn.Linear(G_HIDDEN*2, G_HIDDEN*4), nn.BatchNorm1d(G_HIDDEN*4), nn.ReLU(True),
            nn.Linear(G_HIDDEN*4, IMG_CH*IMG_SIZE*IMG_SIZE),
            nn.Tanh()  # output in [-1, 1]
        )
    def forward(self, z):
        out = self.net(z)
        return out.view(z.size(0), IMG_CH, IMG_SIZE, IMG_SIZE)

class Discriminator(nn.Module):
    # simple MLP discriminator (logit out)
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(IMG_CH*IMG_SIZE*IMG_SIZE, D_HIDDEN), nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(D_HIDDEN, D_HIDDEN*2), nn.LeakyReLU(0.2, inplace=True),
            nn.Linear(D_HIDDEN*2, 1)
        )
    def forward(self, x):
        x = x.view(x.size(0), -1)
        return self.net(x).squeeze(1)

G = Generator().to(device)
D = Discriminator().to(device)

# ----- loss + optim -----
criterion = nn.BCEWithLogitsLoss()             # stable with logits
opt_G = optim.Adam(G.parameters(), lr=LR, betas=BETAS)
opt_D = optim.Adam(D.parameters(), lr=LR, betas=BETAS)

# fixed noise to watch progress
fixed_noise = torch.randn(64, Z_DIM, device=device)

# small helper: visualize + save samples
def show_and_save_samples(epoch, noise=fixed_noise):
    G.eval()
    with torch.no_grad():
        fake = G(noise).cpu()
        grid = vutils.make_grid(fake, nrow=8, normalize=True, value_range=(-1, 1))
    plt.figure(figsize=(6,6))
    plt.axis("off")
    plt.title(f"Generator samples @ epoch {epoch}")
    plt.imshow(grid.permute(1, 2, 0).numpy())
    plt.show()
    # also save to disk for the report
    vutils.save_image(fake, os.path.join(SAMPLES_DIR, f"epoch_{epoch:03d}.png"),
                      nrow=8, normalize=True, value_range=(-1,1))
    G.train()

# ----- training loop -----
g_losses, d_losses = [], []
start_time = time.time()
for epoch in range(1, EPOCHS + 1):
    epoch_g_loss, epoch_d_loss = 0.0, 0.0

    for real, _ in loader:
        real = real.to(device)
        bs = real.size(0)

        # --- Train D: real + fake ---
        opt_D.zero_grad()

        # real batch through D
        real_logits = D(real)
        real_labels = torch.ones(bs, device=device)        # 1 = real
        loss_real = criterion(real_logits, real_labels)

        # fake batch through D (detach so G not updated here)
        z = torch.randn(bs, Z_DIM, device=device)          # noise -> fake
        fake = G(z).detach()
        fake_logits = D(fake)
        fake_labels = torch.zeros(bs, device=device)       # 0 = fake
        loss_fake = criterion(fake_logits, fake_labels)

        d_loss = loss_real + loss_fake                     # total D loss
        d_loss.backward()
        opt_D.step()

        # --- Train G: fool D ---
        opt_G.zero_grad()
        z = torch.randn(bs, Z_DIM, device=device)
        fake = G(z)
        logits = D(fake)
        target_real = torch.ones(bs, device=device)        # want D to say "real"
        g_loss = criterion(logits, target_real)
        g_loss.backward()
        opt_G.step()

        epoch_d_loss += d_loss.item() * bs
        epoch_g_loss += g_loss.item() * bs

    # average per epoch
    epoch_d_loss /= len(loader.dataset)
    epoch_g_loss /= len(loader.dataset)
    d_losses.append(epoch_d_loss); g_losses.append(epoch_g_loss)

    print(f"Epoch [{epoch:3d}/{EPOCHS}]  D_loss: {epoch_d_loss:.4f}  G_loss: {epoch_g_loss:.4f}")

    # save visuals at requested epochs
    if epoch in CHECKPOINT_EPOCHS:
        show_and_save_samples(epoch, fixed_noise)

# ----- loss plots -----
plt.figure(figsize=(7,4))
plt.plot(range(1, EPOCHS+1), d_losses, label="Discriminator loss")
plt.plot(range(1, EPOCHS+1), g_losses, label="Generator loss")
plt.xlabel("Epoch"); plt.ylabel("Loss"); plt.title("GAN training loss (MNIST)")
plt.legend(); plt.grid(True, alpha=0.25)
plt.show()

print(f"Done in {(time.time() - start_time)/60:.1f} min")
print(f"Samples saved to: {os.path.abspath(SAMPLES_DIR)}")
```
